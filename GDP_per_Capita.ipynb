```python
# a) Strip whitespace from 'Country/Territory'

gdp_df['Country/Territory'] = gdp_df['Country/Territory'].str.strip()
```

```python
# Strip leading/trailing whitespace from all string columns
gdp_df = gdp_df.apply(lambda col: col.str.strip() if col.dtype == "object" else col)
```

```python
# b) Convert 'IMF_Year' to datetime
gdp_df['IMF_Year'] = pd.to_datetime(
   gdp_df['IMF_Year'],
   format='%Y',
   errors='coerce'
)
```

```python
gdp_df.head()
```

```python
print(gdp_df['IMF_Year'].dtype)    # should show date time data type
```

```python
#Check the datatype first for another column.
# should be an integer. #Check Dataframe it does show numbers.
print(gdp_df['WorldBank_Estimate'].dtype)
```

```python
# c) Convert 'WorldBank_Estimate' to float with 1 decimal
gdp_df['WorldBank_Estimate'] = (
   pd.to_numeric(gdp_df['WorldBank_Estimate'], errors='coerce')
     .fillna(0)
     .astype(float)
     .round(1)
)
```

```python
print(gdp_df['WorldBank_Estimate'].dtype)
```

```python
#Add on the end for 2 decimal places:
pd.set_option('display.float_format', '{:.2f}'.format)
```

```python
gdp_df.head()
```

```python
# c) Convert 'WorldBank_Estimate' to interger
gdp_df['WorldBank_Estimate'] = pd.to_numeric(gdp_df['WorldBank_Estimate'], errors='coerce').fillna(0).astype(int)
```

```python
gdp_df.head()
```

```python
col = 'IMF_Estimate'
Q1 = gdp_df[col].quantile(0.25)
Q3 = gdp_df[col].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = gdp_df[(gdp_df[col] < lower) | (gdp_df[col] > upper)]
```

```python
import matplotlib.pyplot as plt
numeric_cols = ['IMF_Estimate', 'WorldBank_Estimate', 'UN_Estimate']

for col in numeric_cols:
  plt.figure(figsize=(8, 4))
  plt.boxplot(gdp_df[col].dropna(), vert=False)  # dropna avoids errors from missing values
  plt.title(f'Boxplot of {col} (outliers shown as dots)')
  plt.xlabel(col)
  plt.show()
```

```python
print(f"Number of outliers: {len(outliers)}")
```

```python
print(outliers[['Country/Territory', col]].sort_values(col, ascending=False))
```

```python
print("Outliers shape:", outliers.shape)
```

```python
# To create a Histogram:
plt.figure(figsize=(10, 6))
plt.hist(stu_df['mark'], bins=10, edgecolor='black', color="pink")
plt.title('Distribution of Marks')
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.show()
```

```python
# Line: id vs mark (just to illustrate)
plt.plot(stu_df['id'], stu_df['mark'], color='#9e6999')
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
counts, bins, patches = plt.hist(stu_df['mark'], bins=10, edgecolor='black')
for patch in patches:
   patch.set_facecolor('beige')  # Add this line here
plt.xlabel('Mark')
plt.ylabel('Frequency')
plt.title('Marks histogram with different colours per bin')
plt.show()
```

```python
import numpy as np
counts, bins, patches = plt.hist(stu_df['mark'], bins=10, edgecolor='black')
colors = plt.cm.cool(np.linspace(0, 1, len(patches)))
for patch, color in zip(patches, colors):
   patch.set_facecolor(color)
```

```python
# Count by values in mark column
ax = sns.countplot(x='mark', data=stu_df)
```

```python
#Count of mark
ax = sns.countplot(
   x="mark",
   data=stu_df[stu_df['mark'].notna()],  # fixed: stu_df, not df
   palette="Set3"
)
ax.bar_label(ax.containers[0])
plt.title("Count of Mark")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='class', y='mark', data=stu_df)
plt.title('Marks Distribution by Class')
plt.xlabel('Class')
plt.ylabel('Marks')
plt.show()
```

```python
sns.boxplot(x='class', y='mark', data=stu_df, palette='Set3')
```

```python
plt.figure(figsize=(10, 6))
avg_marks_gender = stu_df.groupby('gender')['mark'].mean().reset_index()
sns.barplot(x='gender', y='mark', data=avg_marks_gender)
plt.title('Average Marks by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Marks')
plt.show()
```
